```python
from matplotlib import pyplot as plt
import torch
import json
```

```python
ids = [f'anew_sl_lif{x}' for x in range(3)] + [f'anew_sl_lif{x}' for x in range(10, 13)]

ids = ['adaptive_smnist', 'cooldown_smnist']
```

```python
to_plot = 'val'
plt.figure(figsize=(10,10))
for id in ids:
    with open(f'../configs/{id}.json', 'r') as config_file:
        config = json.load(config_file)
    data = config['stats'][to_plot]
    val_freq = config['params']['validation_frequency']
    plt.plot(torch.arange(len(data))*val_freq, data, label=id)
#plt.ylim(0.9, 1)
plt.grid()
plt.legend()
plt.xlabel('Iterations')
plt.ylabel('Validation Accuracy')
```
